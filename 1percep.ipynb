```python
#assignment 2 perceptron single layer single discrete
import numpy as np

# Activation function
def step_function(x):
    return np.where(x >= 0, 1, -1)

# Inputs and desired outputs for a single discrete output
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])  # AND gate inputs
D = np.array([-1, -1, -1, 1])  # Desired outputs for the AND gate (-1 and 1 for binary)

# Initialize weights and bias
W = np.random.rand(2)  # Two weights for two input features
b = np.random.rand()   # Bias term
c = 0.1  # Learning rate

# Training loop
for _ in range(1000):  # Number of epochs
    for i, x in enumerate(X):
        output = step_function(np.dot(x, W) + b)  # Compute the perceptron output
        error = D[i] - output  # Calculate error
        W += c * error * x  # Update weights
        b += c * error  # Update bias

# Testing
for x in X:
    output = step_function(np.dot(x, W) + b)
    print(f"Input: {x}, Output: {output}")
```
