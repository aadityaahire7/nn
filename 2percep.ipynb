```python
#perceptron single layer multiple discrete
import numpy as np

# Activation function
def step_function(x):
    return np.where(x >= 0, 1, -1)

# Inputs, desired outputs, and initial weights
X = np.array([[10, 2, 1], [2, -5, 1], [5, 5, 1]])
D = np.array([[1, -1, -1], [-1, 1, -1], [-1, -1, 1]])
W = np.array([[1, -2, 0], [0, -1, 2], [1, 3, -1]], dtype=float)
c = 0.1  # Learning rate

# Training loop
for _ in range(1000):
    for i, x in enumerate(X):
        error = D[i] - step_function(np.dot(x, W))
        W += c * np.outer(x, error)

# Testing
for x in X:
    print(f"Input: {x}, Output: {step_function(np.dot(x, W))}")
```
