```python
def andNotFunc(x1, x2, w1, w2, T):
    results = []
    for i in range(len(x1)):
        sum = x1[i] * w1 + x2[i] * w2
        if sum >= T:
            results.append(0)
        else:
            results.append(1)
    return results

# Inputs
x1 = [1, 1, 0, 0]
x2 = [1, 0, 1, 0]
w1 = w2 = 1
T = 2
and_not_result_T2 = andNotFunc(x1, x2, w1, w2, T)
print("ANDNOT func with T=2:", and_not_result_T2)

def andFunc(x1, x2, w1, w2, T):
    results = []
    for i in range(len(x1)):
        sum = x1[i] * w1 + x2[i] * w2
        if sum >= T:
            results.append(1)
        else:
            results.append(0)
    return results

# Inputs
x1 = [1, 1, 0, 0]
x2 = [1, 0, 1, 0]
w1 = w2 = 1


T = 2
and_result_T2 = andFunc(x1, x2, w1, w2, T)
print("AND func with T=2:", and_result_T2)

def orFunc(x1, x2, w1, w2, T):
    results = []
    for i in range(len(x1)):
        sum = x1[i] * w1 + x2[i] * w2
        if sum >= T:
            results.append(1)
        else:
            results.append(0)
    return results

# Inputs
x1 = [1, 1, 0, 0]
x2 = [1, 0, 1, 0]

# OR gate logic
w1 = w2 = 1
T = 1
or_result_T1 = orFunc(x1, x2, w1, w2, T)
print("OR func with T=1:", or_result_T1)

def threshold(value):
    return 1 if value >= 0 else 0
def signum(value):
    return 1 if value >= 0 else -1
def neuron_1(x1, x2):
    return signum(-2*x1 + x2 - 0.5)

def neuron_2(x1, x2):
    return signum(x1 - x2 - 0.5)

def xor_output(x1, x2):
    n1_output = neuron_1(x1, x2)
    n2_output = neuron_2(x1, x2)
    return threshold(n1_output + n2_output +1)

# Testing the XOR function
for x1 in [0, 1]:
    for x2 in [0, 1]:
        print(f"XOR({x1}, {x2}) = {xor_output(x1, x2)}")

```
