```python
import numpy as np
import matplotlib.pyplot as plt

x_array = np.array([-2, -1, 0, 1, 2])
x_range = np.linspace(-5, 5, 400)

# Sigmoid function Unipolar
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Tanh function Bipolar
def tanh(x):
    exp_x = np.exp(x)
    exp_neg_x = np.exp(-x)
    return (exp_x - exp_neg_x) / (exp_x + exp_neg_x)

# Binary Step function Unipolar
def binary_step(x):
    return np.where(x >= 0, 1, 0)

# Binary Bipolar Step function
def bipolar_step(x):
    return np.where(x >= 0, 1, -1)

# ReLU function Polar
def relu(x):
    return np.maximum(0, x)

# Leaky ReLU function Bipolar
def leaky_relu(x, alpha=0.01):
    return np.where(x >= 0, x, alpha * x)

# Linear activation function
def linear_activation(x):
    return x

# Plotting the functions
plt.figure(figsize=(15, 10))

# Sigmoid
plt.subplot(2, 3, 1)
plt.plot(x_range, sigmoid(x_range))
plt.title('Sigmoid')
plt.grid(True)

# Tanh
plt.subplot(2, 3, 2)
plt.plot(x_range, tanh(x_range))
plt.title('Tanh')
plt.grid(True)

# Binary Step
plt.subplot(2, 3, 3)
plt.plot(x_range, binary_step(x_range))
plt.title('Binary Step')
plt.grid(True)

# Bipolar Step
plt.subplot(2, 3, 4)
plt.plot(x_range, bipolar_step(x_range))
plt.title('Bipolar Step')
plt.grid(True)

# ReLU
plt.subplot(2, 3, 5)
plt.plot(x_range, relu(x_range))
plt.title('ReLU')
plt.grid(True)

# Leaky ReLU
plt.subplot(2, 3, 6)
plt.plot(x_range, leaky_relu(x_range))
plt.title('Leaky ReLU')
plt.grid(True)

plt.tight_layout()
plt.show()

# Linear Activation
plt.figure()
plt.plot(x_range, linear_activation(x_range))
plt.title('Linear Activation')
plt.grid(True)
plt.show()

# Print function outputs for x_array
print("Function outputs for x_array:")
for x in x_array:
    print(f"x = {x}:")
    print(f"  Sigmoid: {sigmoid(x)}")
    print(f"  Tanh: {tanh(x)}")
    print(f"  Binary Step: {binary_step(x)}")
    print(f"  Bipolar Step: {bipolar_step(x)}")
    print(f"  ReLU: {relu(x)}")
    print(f"  Leaky ReLU: {leaky_relu(x)}")
    print(f"  Linear Activation: {linear_activation(x)}")
    print()
```
