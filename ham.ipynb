```python
#hamming
import numpy as np

# Define 10 binary patterns (representing 10 digits)
patterns = np.array([
    [1, 0, 1, 0, 1, 0, 1, 0],  # Pattern for digit 0
    [0, 1, 0, 1, 0, 1, 0, 1],  # Pattern for digit 1
    [1, 1, 0, 0, 1, 1, 0, 0],  # Pattern for digit 2
    [0, 0, 1, 1, 0, 0, 1, 1],  # Pattern for digit 3
    [1, 0, 0, 1, 1, 0, 0, 1],  # Pattern for digit 4
    [0, 1, 1, 0, 0, 1, 1, 0],  # Pattern for digit 5
    [1, 1, 1, 0, 0, 0, 1, 0],  # Pattern for digit 6
    [0, 0, 0, 1, 1, 1, 0, 1],  # Pattern for digit 7
    [1, 0, 1, 1, 0, 0, 1, 1],  # Pattern for digit 8
    [0, 1, 0, 0, 1, 1, 0, 0]   # Pattern for digit 9
])

input_vector = np.array([1, 0, 1, 1, 0, 0, 1, 1])
weights = patterns / 2
activations = np.dot(weights, input_vector) + len(input_vector) / 2

print("Recognized Pattern Index (Digit):", np.argmax(activations)
)
```
