```python
import numpy as np

# Define input values and desired outputs for XOR gate
x = np.array([[1, 1], [1, -1], [-1, 1], [-1, -1]])
y = np.array([-1, 1, 1, -1])
w=np.array([0.,0.])
b=0
eta=1
print(w)
# Hebbian learning algorithm
for i in range(len(x)):
    delta_w = eta * x[i] * y[i]
    delta_b = eta * y[i]
    w += delta_w
    b += delta_b
    print(f"x1: {x[i][0]}, x2: {x[i][1]}, y: {y[i]}, Delta_w1: {delta_w[0]}, Delta_w2: {delta_w[1]}, "
          f"Delta_b: {delta_b}, W1: {w[0]}, W2: {w[1]}, b: {b}")
```
