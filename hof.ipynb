```python
#hopfield
import numpy as np

# Define the pattern to store
pattern = np.array([1, 1, 1, 0])

# Create the weight matrix
weights = np.outer(pattern, pattern)
np.fill_diagonal(weights, 0)  # Set diagonal elements to 0
input_vector = np.array([0, 0, 1, 0])
iterations=5

for i in range(iterations):
    input_vector = np.sign(np.dot(weights, input_vector))
    input_vector[input_vector < 0] = 0  # Convert -1 to 0


print("Corrected Output:", input_vector)
```
